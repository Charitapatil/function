```python
1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
dataset.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset
np.random.seed(0)  # For reproducibility
n = 100  # Number of samples
x = np.random.rand(n) * 100  # Independent variable
y = 2 * x + np.random.normal(0, 10, n)  # Dependent variable with some noise

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', color='blue', alpha=0.7)
plt.title('Scatter Plot of X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

# Display the scatter plot
plt.show()

2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a dataset of random numbers
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=15, size=1000)  # Normal distribution
df = pd.DataFrame(data, columns=['Value'])

# Create the distribution plot
plt.figure(figsize=(10, 6))
sns.histplot(df['Value'], bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Numerical Variable')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid()

# Display the distribution plot
plt.show()

3. Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset representing categories and their corresponding values
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [25, 40, 35, 50, 30]
}
df = pd.DataFrame(data)

# Create a bar plot to compare categories
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Category', y='Values', palette='viridis')
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(axis='y')

# Display the bar plot
plt.show()

4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a dataset with categories and numerical values
np.random.seed(0)  # For reproducibility
categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, size=200),  # Randomly assign categories
    'Value': np.random.normal(loc=50, scale=15, size=200)  # Normally distributed values
}
df = pd.DataFrame(data)

# Create a box plot to visualize the distribution of values across categories
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Value', palette='Set2')
plt.title('Distribution of Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y')

# Display the box plot
plt.show()


5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(0)  # For reproducibility
n_samples = 100
# Create features with some correlation
x1 = np.random.normal(0, 1, n_samples)
x2 = x1 * 0.5 + np.random.normal(0, 0.5, n_samples)  # x2 is correlated with x1
x3 = x1 * -0.3 + np.random.normal(0, 0.5, n_samples)  # x3 is negatively correlated with x1
x4 = np.random.normal(0, 1, n_samples)  # x4 is independent

# Create a DataFrame
df = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3, 'Feature 4': x4})

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix Heatmap')
plt.show()
```
